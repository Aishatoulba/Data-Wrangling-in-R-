## 1.Tidy Data

### 1. What is tidy data ?
Tidy Data describes data that's been put into a standardized format that facilitates future analytic work. Hadley Wickham, a data scientist who is one of the key developers of the R language coined the term Tidy Data in this paper that he published in the Journal of Statistical Software back in 2014.

Converting data from its original format into Tidy form is difficult, time-consuming work, but why would we want to spend this time and effort required to create Tidy Data? Well, there were three reasons:
* First, Tidy Data facilitates initial exploration and analysis. If our data is in a standardized format, it's much easier to notice trends, anomalies and other important features of datasets.

* Tidy Data also improves our ability to collaborate with others. If our data is in standard formats, we can easily share it with other people who will then be able to quickly begin analyzing it without having to go through their own data wrangling work first.

* Finally, if we convert our data to Tidy format, we can take advantage of many R packages that accept Tidy Data as input without performing additional transformations.

### 2. Variables,observations and values
Tidy data uses three terms to describe the data that we structure. 

* Variables measure some underlying attribute of the entity that we're measuring. For example, if we're measuring people, variables might be their height, weight and blood pressure.

* Observations are the collections of all of the variable measurements related to a single entity, normally at a single point in time. For example, we might record a person's height, weight, and blood pressure on Tuesday. That would be a single observation. If we measured that same person's vital statistics again on Thursday, that would be a second observation.

* Values are the intersection of a variable and an observation. For example, the value of my height on Tuesday might be 70 inches.

### 3. Common data problems
The common problems that appear in untidy datasets are : 
- where the column header contain values instead of variable names.
- datasets that store multiple variables in a single column.
- datasets that store variables in both rows and columns.
- datasets that store different types of observational units in the same table.
- datasets where a single observational unit is spread across multiple tables.

### 4. what is Tidyverse

The Tidyverse is a collection of R packages that help you easily work with tidy data.The most important components of the Tidyverse are : 
* The Tibble package provides you with the tools that you'll need to replace R's concept of data frames with a new data structure called Tibbles that form the core of the Tidyverse. 
* The readr and readxl packages provide you with the ability to easily import and export data into R. 
* Dplyr and tidyr are data manipulation libraries that contain many of the functions that we'll use to resolve issues with datasets and convert them into tidy form.
*  Stringr is a string manipulation library that makes it easy to work with text while lubridate does the same thing for dates and times.
* ggplot2 is a data visualization library that produces charts and other data visualizations that are easy to create and interpret. 

## 2.Working with tibbles

### 1. Building and printing tibbles

Tibbles are like core data structure of the tidyverse. And we use them to facilitate the display and analysis of information in tidy format.
Although data frames are the most common data structure used to store complex data sets in base R , tibbles have a few advantages over data frames that make them much easier to use:

* First, all tidyverse packages are designed to use tibbles for both input and output where applicable. This standardized data format makes it easy for our developers to string together functions from different packages, without having to perform intermediate data conversions.

* Second, tibbles include functionality that makes it much easier to print and display output than with basic data frames.

* Third, tibbles don't make the assumptions that you commonly find when creating data frames. For instance dataframes often convert character strings to factos and we often have to override the setting. Tiblle don't try to make those conversions automatically.


There are 3 ways to build tibbles : 
* as_tibble() builds a tibble from an existing data frame.
* tibble() builds a tibble from scratch. 
* Import builds a tibble from an external data source.

```r
library(tidyverse)
CO2_tibble <- as_tibble(CO2)
```

```r
# building tibble from scratch 
names <- c("Mike","Matt","Chris","Ricky")
birth_year <- c("1996","2000","1972","2003")
eyecolor <- c("blue","black","black","blue")

people <- tibble(names,birth_year,eyecolor)
head(people)
```

### 2. Subsetting tibbles
Subsetting consists of extracting a single variable from a tibble for further analysis.There are two ways of subsetting : 
* using a dollar sign $ 
* [[]] operator which extracts a variable as vector from tibble by name or position.

```r
# using $ 
people$eyecolor
# using [[]] with column name
people[['eyecolor']]
# using [[]] with column index
people[[3]]
```

### 3. Filtering tibbles
Filtering provides us with a way to reduce the number of rows in your tibble. When performing filtering, we specify some criteria that are used to reduce the number of rows in a dataset.

```r
# people with black eyes 
filter(people,eyecolor=='black')
# people born after 1972 
filter(people,birth_year > 1972)
# people with blue eyes born after 1972 
filter(people,eyecolor=='blue' & birth_year >1972)
```

## 4.Data Transformation

### 1. Wide vs long datasets
The difference between wide and long datasets boils down to whether we prefer to have more columns in our dataset or more rows. A dataset that emphasizes putting additional data about a single subject in columns is called a wide dataset because, as we add more columns, the dataset becomes wider. Similarly, a dataset that emphasizes including additional data about a subject in rows is called a long dataset because, as we add more rows, the dataset becomes longer. It's important to point out that there's nothing inherently good or bad about wide or long data.


### 2. Making wide datasets long : pivot_longer()

```r
library(tidyverse)
library(readr)
pew <- read_csv("C:/Users/Atolb/pew.csv")
```

### 1. Detecting outliers
Outliers are data points that lie far outside the norm, and they may occur for two reasons. First, outliers may indicate some type of error in the data set. Someone may have measured the data incorrectly in the first place, incorrectly input it into a system or performed a calculation improperly. 

Not all outliers are errors however. They may be real data points that result from a skewed dataset. For example, imagine that you have a dataset consisting of the salaries of corporate executives. You might find that most have salaries in the hundreds of thousands of dollars, but then there's a single data point of 1 dollar . After investigating, you might discover that a corporate CEO agreed to take a $1 salary due to the company's poor performance that year. That's a correct data point and it should likely be retained in the dataset.

We can use a variety of visual tools to identify outliers such as scatterplots , histograms and boxplots.

* Histograms : They group data points into bins representing different numeric ranges.it's basically a bar chart that works across continuous variables instead of categorical variables.

* Boxplots : The second graphical tool that we can use is called the boxplot. This is an interesting tool used by statisticians to help identify outliers.

   - Minimum Score : The lowest score, excluding outliers (shown at the end of the left whisker).

   - Lower Quartile: Twenty-five percent of scores fall below the lower quartile value (also known as the first quartile).

   - Median : The median marks the mid-point of the data and is shown by the line that divides the box into two parts (sometimes known as the second quartile). Half the scores are greater than or equal to this value, and half are less.
   
   - Upper Quartile : Seventy-five percent of the scores fall below the upper quartile value (also known as the third quartile). Thus, 25 % of data are above this value.

  - Maximum Score : The highest score, excluding outliers (shown at the end of the right whisker).
